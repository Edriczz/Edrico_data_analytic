```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
import os
```

```python
from google.colab.patches import cv2_imshow
```

```python

# --- Pengaturan ---
# Ganti dengan path video dan folder database Anda
video_path = "/content/drive/MyDrive/Face Recognition Test/ObamaBiden.mp4"
db_path = "/content/drive/MyDrive/Face Recognition Test/Foto"
output_path= "facerec.mp4"
# Buka file video
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Tidak bisa membuka file video.")
    exit()
# BARU: Dapatkan properti video untuk VideoWriter
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# BARU: Inisialisasi VideoWriter
# 'mp4v' adalah kode FourCC untuk format .mp4
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
writer = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))


# Hitung frame untuk dilewati
frame_skip = 5
frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Proses hanya frame tertentu untuk kecepatan
    frame_count += 1
    if frame_count % frame_skip == 0:
        try:
            # Cari wajah di frame
            dfs = DeepFace.find(img_path=frame,
                                 db_path=db_path,
                                 enforce_detection=False,
                                 model_name='VGG-Face'
                                )

            # Loop melalui setiap wajah yang ditemukan
            for df in dfs:
                if not df.empty:
                    best_match = df.iloc[0]
                    identity_path = best_match['identity']

                    name = os.path.splitext(os.path.basename(identity_path))[0]

                    x, y, w, h = best_match['source_x'], best_match['source_y'], best_match['source_w'], best_match['source_h']

                    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
                    cv2.putText(frame, name, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        except Exception as e:
            pass # Lanjutkan jika ada error di frame ini

    # DIUBAH: Tulis frame ke file video output, bukan menampilkannya
    writer.write(frame)

print(f"Video selesai diproses dan disimpan sebagai {output_path}")

# Lepaskan video dan writer
cap.release()
writer.release() # BARU: Jangan lupa release writer
```
